```python
# Function for Open Chrome and Login to LinkedIN
def open_chrome():
    chrome_options = Options()
    chrome_options.add_argument("--start-maximized")  # Buka browser dalam mode layar penuh

    driver = webdriver.Chrome(options=chrome_options)

    # Buka LinkedIn
    driver.get("https://www.linkedin.com/login")

    # Tunggu beberapa detik untuk memastikan halaman terbuka
    time.sleep(3)

    email_input = driver.find_element(By.XPATH, "//input[@aria-label='Email or phone']")
    email_input.send_keys("compaic3@gmail.com")

    password_input = driver.find_element(By.XPATH, "//input[@id='password']")
    password_input.send_keys("Compaic123456-,")

    # Tekan tombol login
    password_input.send_keys(Keys.RETURN)
    
    return driver
    # End Function for Open Chrome and Login to LinkedIN

def open_profile(driver):
    # Buka LinkedIn
    driver.get("https://www.linkedin.com/in/zulfi-fadilah-azhar/")
    try:
        popup = driver.find_element(By.XPATH, "//section[@aria-labelledby='base-contextual-sign-in-modal-modal-header']")
        popup = popup.find_element(By.XPATH, "//button[@data-tracking-control-name='public_profile_contextual-sign-in-modal_modal_dismiss']")
        popup.click()
    except Exception as e :
        print("Gaada Popup")

    time.sleep(5)
    # Cari elemen <div> yang mengandung teks 'skills' di dalamnya
    # skill_click = driver.find_element(By.XPATH, "//div[@class='pvs-list__footer-wrapper' and contains(., 'skills')]")
    skill_click = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//div[@class='pvs-list__footer-wrapper' and contains(., 'skills')]"))
    )
    # Klik elemen tersebut (jika klik pada elemen <div> itu diperlukan)
    skill_click.click()
    return driver

def get_skill(driver, skills) :
    time.sleep(1.5)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    skill_ul = WebDriverWait(driver, 2).until(
        EC.presence_of_element_located((By.XPATH, "//ul[@class='MyAlsvVeiPoYBRwzMGJPPEPCPewbSzASrEg ']"))
    ) 
    # Temukan semua elemen <li> di dalam <ul> tersebut
    # Cari semua elemen <li> yang memiliki class tertentu
    li_elements = skill_ul.find_elements(By.XPATH, '''//div[@class="display-flex flex-wrap align-items-center full-height"]''')

    # Ambil teks dari setiap elemen yang ditemukan
    for li in li_elements:
        teks = li.text
        if teks != "":
            skills.append(teks.split('\n')[0])
        else:
            break

    driver.back()
    return driver

def get_licences(driver, licenses) : 
    driver.execute_script("window.scrollBy(0, -700);")
    lisense_click = WebDriverWait(driver, 2).until(
        EC.presence_of_element_located((By.XPATH, "//div[@class='pvs-list__footer-wrapper' and contains(., 'licenses & certifications')]"))
    )
    lisense_click.click()
    
    time.sleep(1)

    # Scroll ke bawah sampai ke bagian paling bawah
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    license_div = license_ul = WebDriverWait(driver, 3).until(
        EC.presence_of_element_located((By.XPATH, "//div[@class='pvs-list__container']"))
    ) 

    license_ul = license_div.find_element(By.XPATH, "//ul[@class='MyAlsvVeiPoYBRwzMGJPPEPCPewbSzASrEg ']") 

    li_elements = license_ul.find_elements(By.XPATH, '''//div[@class="display-flex flex-wrap align-items-center full-height"]''')

    # Ambil teks dari setiap elemen yang ditemukan
    for li in li_elements:
        teks = li.text
        teks = teks.split('\n')
        if len(teks) == 2:
            print(teks[0])
            licenses.append(teks[0])
        else:
            break

    driver.back()
    return driver



skills = []
licenses = []
experiences = []
driver = open_chrome()
driver = open_profile(driver)
driver = get_skill(driver, skills)
get_licences(driver, licenses)

```
